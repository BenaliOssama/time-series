```python
import pandas as pd
```

```python
# pd.date_range generates a sequence of dates
# freq='D' means one entry per Day
index = pd.date_range(start='2010-01-01', end='2020-12-31', freq='D')
index
```

```python
# range(len(index)) gives us 0, 1, 2, ... up to however many dates there are
integer_series = pd.Series(data=range(len(index)), index=index, name='integer_series')
integer_series
```

```python
# .rolling(7) says "look at a window of 7 days"
# .mean() says "compute the average within that window"
rolling_avg = integer_series.rolling(7).mean()

rolling_avg
#rolling_avg.tail(10)
```
